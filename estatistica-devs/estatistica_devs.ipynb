```python
import pandas as pd
```

```python
pd.__version__
```

```python
dict_medidas = {
    "idade": [15, 18, 25, 25, 40, 55, 58, 60, 80],
    "altura": [160, 162, 165, 168, 172, 174, 174, 174, 176],
}
```

```python
dict_medidas
```

```python
df_medidas = pd.DataFrame.from_dict(dict_medidas)
```

```python
df_medidas
```

# Medidas Resumo

### Medidas de Posição

```python
# media
df_medidas["idade"].mean()
```

```python
# mediana
df_medidas["idade"].median()
```

```python
# moda
df_medidas["idade"].mode()
```

```python
# moda altura
df_medidas["altura"].mode()
```

### Medidas de Dispersão

```python
# variancia
# sintaxe alternativa
df_medidas.idade.var() # diferença entre a média e cada valor, elevado ao quadrado
```

```python
# desvio padrão
df_medidas.idade.std() # raiz quadrada da variância
```

```python
# coeficiente de variação -idade (desvio padrão dividido pela média, multiplicado por 100 para obter o resultado em porcentagem)
df_medidas.idade.std() / df_medidas.idade.mean() * 100
```

```python
# coeficiente de variação - altura
df_medidas.altura.std() / df_medidas.altura.mean() * 100
```

### Medidas de Forma

```python
# Assimetria - Idade
df_medidas.idade.skew()
```

```python
# Assimetria - Altura
df_medidas.altura.skew()
```

```python
# Curtose - Idade
df_medidas.idade.kurtosis()
```

```python
# Curtose - Altura
df_medidas.altura.kurtosis()
```

```python
# Obter medidas estatísticas 
df_medidas.idade.describe()
```

### Correlação de variáveis

```python
# Correlação do DataFrame, segundo Pearson
df_medidas.corr(method='pearson')
```

```python
# Correlação entre duas variáveis
df_medidas.idade.corr(df_medidas.altura, method='pearson')
```

```python
# Correlação de DataFrame, segundo Spearman
df_medidas.corr(method='spearman')
```
